```python
labelencoder = preprocessing.LabelEncoder()
lbs=['Ship Mode',
'Segment',
'City',
'State',
'Country',
'Market'
,'Region'
,'Category'
,'Sub-Category'
,'Order Priority'
,'order_day_of_week']
```

```python
def extract_date(data , column):
    for columns in column:
        data[columns+'_year']=data[columns].apply(lambda x:x.year)
        data[columns+'_month']=data[columns].apply(lambda x:x.month)
        data[columns+'_date']=data[columns].apply(lambda x:x.date)
```

```python
extract_date(data , ['Order Date' , 'Ship Date'])
```
